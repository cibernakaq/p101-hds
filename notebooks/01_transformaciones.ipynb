### Creamos los campos `item_id`, `recommend` y `review`.

```python
def helper_fn(list_):
  recommends = []
  for item in list_:
    recommends.append(item['recommend'])
  return recommends

reviews_df['recommend'] = reviews_df.reviews.map(helper_fn)
```

```python
np.random.seed(42)
reviews_df.recommend.sample(n=10)
```

```python
def helper_fn2(list_):
  reviews = []
  for item in list_:
    reviews.append(item['review'])
  return reviews

reviews_df['review'] = reviews_df.reviews.map(helper_fn2)
```

```python
np.random.seed(42)
reviews_df.review.sample(n=10)
```

```python
def helper_fn3(list_):
  item_ids = []
  for item in list_:
    item_ids.append(item['item_id'])
  return item_ids

reviews_df['item_id'] = reviews_df.reviews.map(helper_fn3)
```

```python
np.random.seed(42)
reviews_df.item_id.sample(n=10)
```

### Creamos los campos `item_id`, `item_name`, `playtime_forever`.

```python
def helper_fn(list_):
  item_ids = []
  for item in list_:
    item_ids.append(item['item_id'])
  return item_ids

items_df['item_id'] = items_df['items'].map(helper_fn)
items_df['item_id'].head()
```

```python
def helper_fn(list_):
  item_playtimes = []
  for item in list_:
    item_playtimes.append(item['playtime_forever'])
  return item_playtimes

items_df['item_playtime'] = items_df['items'].map(helper_fn)
items_df['item_playtime'].head()
```

```python
def helper_fn(list_):
  item_names = []
  for item in list_:
    item_names.append(item['item_name'])
  return item_names

items_df['item_name'] = items_df['items'].map(helper_fn)
items_df['item_name'].head()
```
